```python
class ImageMatrixHandler:

    def get_mv_data(json_file_path):
        with open(json_file_path,'r') as file:
            global json_data
            json_data = json.load(file)
    
    def generate_np_marix_from_json(mouse_movement_data):
        all_mv = []
        list_mv = []
        mv_string = mouse_movement_data 

        len_mv = len(mv_string)
        i=0
        while i<len_mv:
            while i<len_mv and  mv_string[i] != '(':
                # print(mv_string[i])
                i+=1
            j = i
            while j<len_mv and mv_string[j] != ')':
                # print(mv_string[j])
                j+=1

            try:
                val = eval(mv_string[i:j+1])
                if isinstance(val ,tuple):
                    list_mv.append(val)
                if(i>2 and mv_string[i-2]=='c' and mv_string[i-1]=='l'):
                    all_mv.append(list_mv.copy())
                    list_mv = []
            except NameError:
                temp = [x for x in list_mv]
                # print(temp[:20])
                all_mv.append(temp)
                list_mv = []
            except SyntaxError:
                print("$",mv_string[i:j+1])
            i = j+1
        if len(list_mv)>0:
            all_mv.append(list_mv.copy())
        len_mv
        len(all_mv)

        arrays = {}
        arr_num = 0
        for points in all_mv:
            arrays[f"mv_np_arr_{arr_num}"] = ImageMatrixHandler.get_2d_np_plot(points)
            arr_num += 1


        # Save all arrays in one file
        return arrays

    def save_data_for_phase1_mv():
        data_path = ["../dataset/phase1/data/mouse_movements/humans_and_advanced_bots","../dataset/phase1/data/mouse_movements/humans_and_moderate_bots"]
       
        for data_dir in data_path:
            for dir in os.listdir(data_dir):
                for file in os.listdir(os.path.join(data_dir,dir)):
                    json_data = None
                    with open(os.path.join(data_dir,dir,file),'r') as fd:
                        json_data = json.load(fd)

                    arrays = ImageMatrixHandler.generate_np_marix_from_json(json_data["total_behaviour"])
                    mv_data_dir = "mv_data/" 

                    if os.path.exists(mv_data_dir + dir):
                            print("path exist for ",dir)
                    np.savez(mv_data_dir + dir + ".npz", **arrays)
                print(".",end="")
            print("\n done for :" ,data_dir)

        print("done for all")

    def save_data_for_phase2_mv():
        data_path = ["../dataset/phase2/data/mouse_movements/bots","../dataset/phase2/data/mouse_movements/humans"]

        for dir in data_path:
            for file in os.listdir(dir):
                with open(os.path.join(dir,file),'r') as fd:
                    for json_text_line in fd:
                        json_data = json.loads(json_text_line)

                        arrays = ImageMatrixHandler.generate_np_marix_from_json(json_data["mousemove_total_behaviour"])
                        data_dir = "mv_data2/" 
                        if os.path.exists(data_dir + json_data["session_id"] + ".npz"):
                            print("path exist for ",json_data["session_id"])
                        np.savez(data_dir + json_data["session_id"] + ".npz", **arrays)

                        print(".",end="")
                print("\n done for :" ,file)

        print("done for all")


        # Load them back

    def load_mv_data(file_path = "data.npz"):
        data = np.load(file_path)
        return data

    def get_2d_np_plot(points):
        # x_coords = [p[0] for p in points]
        # y_coords = [p[1] for p in points]

        # # Plotting the points
        # plt.plot(x_coords, y_coords, 'o-') # 'o-' plots markers and connects them with a line
        # # Or, to plot only markers:
        # # plt.plot(x_coords, y_coords, 'o') 

        # plt.title('2D Points Plot (using plot)')
        # plt.xlabel('X-axis')
        # plt.ylabel('Y-axis')
        # plt.show()

        x_coords = [p[0] for p in points]
        y_coords = [p[1] for p in points]

        # Create plot (but don't show it on screen)
        fig, ax = plt.subplots(figsize=(72, 48), dpi=10)
        ax.plot(x_coords, y_coords, linewidth=15)  # 'o-' = markers + connecting lines
        ax.axis('off')
        # ax.set_title('2D Points Plot')
        # ax.set_xlabel('X-axis')
        # ax.set_ylabel('Y-axis')

        # Render the figure
        fig.canvas.draw()

        # Convert to RGB array
        width, height = fig.canvas.get_width_height()
        rgb = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
        rgb = rgb.reshape((height, width, 3))

        # Convert to grayscale (luminance formula)
        gray = np.dot(rgb[..., :3], [0.2989, 0.5870, 0.1140])
        gray = gray / 255.0  # normalize to 0â€“1

        plt.close(fig)  # close figure to save memory

        # print("Grayscale image shape:", gray.shape)
        # print("Example pixel value:", gray[100, 100])

        # plt.imshow(gray, cmap='gray')   # 'gray' colormap forces grayscale display
        # plt.axis('off')                 # hide axis for clean view
        # plt.show()

        return gray




```
